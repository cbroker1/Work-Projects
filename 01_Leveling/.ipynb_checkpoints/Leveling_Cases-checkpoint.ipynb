```python
import pandas as pd
import numpy as np
import datetime as dt
#open csv file
df = pd.read_csv(r'C:\Users\cbroker1\Desktop\CASE1.csv', encoding = 'unicode_escape')
```

```python
#Phase 1 Cleaning: 

#Converting 'Duration' decimal
#df['Duration'] = df['Duration'].replace('0','0:00')
#df['Duration'] = pd.DatetimeIndex(df['Duration']).hour + pd.DatetimeIndex(df['Duration']).minute/60

#Cleaning: Remove "The" from Institution Column
df['Institution'] = df['Institution'].replace('The Johns Hopkins Medical Institutions','Johns Hopkins Medical Institutions')
df['Institution'] = df['Institution'].replace('The Johns Hopkins University','Johns Hopkins University')

#Cleaning: Remove 'Family of Faculty' and replace with 'Faculty'
df['Affiliation'] = df['Affiliation'].replace('Family of Faculty', 'Faculty')

#Cleaning: Remove 'Family of Staff' with 'Staff'
df['Affiliation'] = df['Affiliation'].replace('Family of Staff', 'Staff')

#Cleaning: Remove 'Student??SOM Graduate' and 'Family of Student' with 'Student'
df['Affiliation'] = df['Affiliation'].replace(['Student Âƒ?? SOM Graduate', 'Family of Student'], 'Student')

#Cleaning: Remove 'Johns Hopkins Com. Physicians' with 'Johns Hopkins Communtiy Physicians'
df['Division'] = df['Division'].replace('Johns Hopkins Com. Physicians', 'Johns Hopkins Community Physicians')

#Cleaning: Remove 'Bloomberg School of Public Health' with 'School of Public Health'
df['Division'] = df['Division'].replace('Bloomberg School of Public Health', 'School of Public Health')

#Cleaning: Remove 'Public Health NonCredit' with 'School of Public Health'
df['Division'] = df['Division'].replace('Public Health NonCredit', 'School of Public Health')

#Cleaning: Remove 'Nitze School of Advanced International Studies' with 'SAIS'
df['Division'] = df['Division'].replace('Nitze School of Advanced International Studies', 'SAIS')

#Cleaning: Remove 'School of Arts and Sciences' with 'Krieger School of Arts and Sciences'
df['Division'] = df['Division'].replace('School of Arts & Sciences', 'Krieger School of Arts and Sciences')

#Cleaning: Remove 'JHU Research & Clinical Fellow', 'etc' with 'Postdoc'
df['Job Title'] = df['Job Title'].replace(['JHU Clinical Fellow',
                                   'JHU Research & Clinical Fellow',
                                   'POST DOC FELLOW',
                                   'POST DOCTORAL FELLOW',
                                   'POSTDOC FELLOW',
                                   'Postdoctoral Fellow',
                                   'POSTDOCTORAL FELLOW',
                                   'Postdoctoral Fellow',
                                   'POSTDOCTORAL FELLOW',
                                   'PostDoc',
                                   'JHU Research Fellow'],
                                   'Postdoc')

#Cleaning: Remove 'Resident/Intern Terms' with 'SOM Housestaff'
df['Job Title'] = df['Job Title'].replace(['Fellow/JHH Intern',
                                   'Fellow/Bayview Assistant Resident',
                                   'Fellow/JHH Assistant Resident'],
                                   'SOM Housestaff')
```

```python
#Insertion of Blank Columns
df['Level_1'] = np.nan
df['Level_2'] = np.nan
df['Level_3'] = np.nan
```
